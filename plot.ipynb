```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("trainingSet_final.csv")
Y = data["PrecipitationSumInches"]
```

```python
x_vis = data.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent',
                  'SeaLevelPressureAvgInches', 'VisibilityAvgMiles',
                  'WindAvgMPH'], axis = 1)
# print(x_vis)

plt.scatter(x_vis[x_vis.columns.values[0][:100]],Y,color = 'g')
plt.title(x_vis.columns.values[0])
plt.show()
```

```python
for i in range(1,6):
    plt.scatter(x_vis[x_vis.columns.values[i][:100]],Y,color = 'g')
    plt.title(x_vis.columns.values[i])
    plt.show()
```

```python
days = [ i for i in range(Y.size)]
```

```python
for i in range(0,6):
    plt.scatter(days,x_vis[x_vis.columns.values[i][:100]],color = 'g')
    plt.title(x_vis.columns.values[i])
    plt.show()
```
